## Exercise 1 â€” Most Frequent Characters

Write a function `most_frequent(s)` that:
- counts how many times each character appears
- stores results as a **list of (character, count) tuples**
- prints the characters in decreasing order of frequency


```python
def most_frequent(s):
    lst = []
    letters = []
    for ch in s:
        if ch not in letters:
            letters.append(ch)

    for letter in letters:
        lst.append((letter, s.count(letter)))
    while lst:
        max_count = 0
        max_item = None
        for item in lst:
            if item[1] > max_count:
                max_count = item[1]
                max_item = item

        print(max_item)
        lst.remove(max_item)

# Test string
most_frequent('banana')
```

## Exercise 2 â€” Word Frequencies

Given a list of words, count how often each word appears.
Return a list of `(word, count)` tuples sorted by count (descending).


```python
def word_frequencies(words):
    # TODO
    lst = []
    lst2 = []
    for word in words:
      if word not in lst:
        lst.append(word)
    for word in lst:
      lst2.append((word, words.count(word)))
    return sorted(lst2, reverse = True)





words = ['the', 'cat', 'the', 'dog']
print(word_frequencies(words))
```
